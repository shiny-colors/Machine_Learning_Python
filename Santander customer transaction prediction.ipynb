```python
#
y = np.array(df["target"], dtype="int")
N = y.shape[0]
```

```python
#
ones = np.repeat(1, N).reshape(N, 1)
x = np.hstack((ones, np.array(df.iloc[:, 3:])))
k = x.shape[1]
```

```python
########
##
def define_likelihood(beta, y, X, N):
    #
    mu = np.exp(np.dot(X, beta))
    Pr = mu / (1 + mu)
    
    #
    LLi = y * np.log(Pr) + (1-y) * np.log(1-Pr)
    LL = -sum(LLi)
    return LL
```

```python
##
def dloglike(beta, y, x, N):
    #
    mu = np.exp(np.dot(x, beta))   #
    Prob = mu / (1 + mu)   #

    #
    dlogit = y.reshape(N, 1)*x - Prob.reshape(N, 1)*x

    #
    LLd = -np.sum(dlogit, axis=0)
    return(LLd)
```

```python
theta0 = np.zeros(k)
res = optimize.minimize(define_likelihood, theta0, jac=dloglike, method='BFGS', args=(y, x, N), 
                        options={"gtol": 0.1, "disp":True})
```

```python
##
#
theta = res.x   #
hessian = res.hess_inv   #
LL = -res.fun   #
```

```python
t_value = theta / np.sqrt(np.diag(hessian))   #t
AIC = -2*LL + 2*(k+1)   #AIC
BIC = -2*LL + np.log(N)*(k+1)   #BIC
print(np.round(t_value, 3))
print(np.round(np.array([LL, AIC, BIC]), 3))
```
